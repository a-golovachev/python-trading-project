```python
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
def gen_returns(df):
    df['r_1d'] = ((df['ADJ_CLOSE'].shift(-1) - df['ADJ_CLOSE']) / df['ADJ_CLOSE'])
    df['r_1w'] = ((df['ADJ_CLOSE'].shift(-5) - df['ADJ_CLOSE']) / df['ADJ_CLOSE'])
    df['r_1m'] = ((df['ADJ_CLOSE'].shift(-22) - df['ADJ_CLOSE']) / df['ADJ_CLOSE'])
    df['log_r_1d'] = np.log(1 + df['r_1d'])
    df['log_r_1w'] = np.log(1 + df['r_1w'])
    df['log_r_1m'] = np.log(1 + df['r_1m'])
```

```python
rosn = pd.read_csv('ROSN_quotes.csv')
lkoh = pd.read_csv('LKOH_quotes.csv')
sibn = pd.read_csv('SIBN_quotes.csv')
imoex = pd.read_csv('IMOEX_quotes.csv')
mcftr = pd.read_csv('MCFTR_quotes.csv')
usd = pd.read_excel('USDRUB.xlsx')
ruonia = pd.read_excel('RUONIA.xlsx')
```
